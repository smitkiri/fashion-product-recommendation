# Loading Test and Train Images

```python
train_images = utils.open_pickle(file = PICKLE_PATH + "train_images" )
train_images_idx = utils.open_pickle(file = PICKLE_PATH + "train_images_names")

test_images = utils.open_pickle(file = PICKLE_PATH + "test_images")
test_images_idx = utils.open_pickle(file = PICKLE_PATH + "test_images_names")
```

```python
IMG_DIMENSION = train_images[0].shape
```

```python
IMG_DIMENSION
```

```python
vgg16 = VGG16(weights = 'imagenet', input_shape = IMG_DIMENSION, include_top = False)
vgg16.trainable = False
```

```python
train_images_embd = []
total_train = len(train_images)
for i in range(total_train):
    
    train_images_embd.append(utils.get_embeddings(image=train_images[i], 
                                                  dimension=IMG_DIMENSION, model=vgg16).flatten())
    if (i+1)%100==0: print("Generated embeddings for %d out of %d"%(i+1, total_train))
train_images_embd = np.array(train_images_embd)
```

```python
test_images_embd = []
total_test = len(test_images)
for i in range(total_test):
    
    test_images_embd.append(utils.get_embeddings(image=test_images[i], 
                                                  dimension=IMG_DIMENSION, model=vgg16).flatten())
    if (i+1)%100==0: print("Generated embeddings for %d out of %d"%(i+1, total_test))
test_images_embd = np.array(test_images_embd)
```

# Saving Train and Test Images Embeddings

```python
utils.save_pickle(PICKLE_PATH + "train_images_embd", train_images_embd)
utils.save_pickle(PICKLE_PATH + "test_images_embd", test_images_embd)
```
